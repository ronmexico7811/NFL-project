```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
plt.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')
```

```python
import os 

os.chdir('/Users/corbettcarrel/NFL-project/Data')
```

```python
combine = pd.read_csv('combine_data_since_2000_PROCESSED_2018-04-26.csv')
combine.head()
```

```python
combine.describe()
```

```python
combine.columns
combine_slim = combine[['Player', 
                        'Pos', 
                        'Ht', 
                        'Wt', 
                        'Forty', 
                        'Vertical', 
                        'BenchReps',
                        'BroadJump', 
                        'Cone', 
                        'Shuttle', 
                        'Year', 
                        'AV',  
                        'Round']]
combine_slim = combine_slim.dropna()
print(combine_slim.info())
combine_slim.head()
```

```python
dummy = pd.get_dummies(combine_slim['Pos'])
combine_clean = pd.concat([combine_slim, dummy], axis = 1)
combine_pos = combine_clean[['Player',
                              'C', 
                              'CB', 
                              'DE',
                              'DT', 
                              'FB', 
                              'FS', 
                              'ILB', 
                              'LS', 
                              'OG', 
                              'OLB', 
                              'OT', 
                              'QB', 
                              'RB', 
                              'SS',
                              'TE', 
                              'WR',
                              'Round']]

print(combine_pos.info())
combine_pos.head()
```

```python
combine_slim.corr()
```

```python
#heatmap
# calculate and visualize correlations as matrix
plt.style.use('default')

combine_corr = combine_slim.corr()

# select a blue color map and define vmin
heatmap = plt.pcolor(combine_corr, cmap='RdBu', vmin=-1)

# plot the legend on the side
plt.colorbar(heatmap)

# Define the tick marks so that they can be easily read
plt.xticks(np.arange(len(combine_corr.columns)), combine_corr.columns, rotation='vertical');
plt.yticks(np.arange(len(combine_corr.columns)), combine_corr.columns, rotation='horizontal');
```
